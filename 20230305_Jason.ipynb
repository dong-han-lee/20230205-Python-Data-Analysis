## 關於回測

```python
!pip install backtesting
```

```python
from backtesting import Backtest, Strategy
from backtesting.lib import crossover

from backtesting.test import SMA, GOOG


class SmaCross(Strategy):
    def init(self):
        price = self.data.Close
        self.ma1 = self.I(SMA, price, 10)
        self.ma2 = self.I(SMA, price, 20)

    def next(self):
        if crossover(self.ma1, self.ma2):
            self.buy()
        elif crossover(self.ma2, self.ma1):
            self.sell()

```

```python
bt = Backtest(TSMC.tail(100), SmaCross, commission=.002, exclusive_orders=True)
stats = bt.run()
bt.plot()
```

### 抓取00878歷史股價

```python
HDV = yf.download('00878.tw')
```

```python
HDV.head()
```

```python
HDV['Close'].tail(100).plot()
```

```python
# 假設每一次買五萬元
CASH = 50000
```

```python
# 假設定期定額的時限是2022年一整年
for i in range(12):
  print(f"2022-{i+1:02}")
```

```python
# 總共花的錢
TOTALCASH = CASH*12
```

```python
# 總共買的股數
TOTALSTOCK = 0
for i in range(12):
  d = f"2022-{i+1:02}"
  TOTALSTOCK += CASH/HDV[d]['Close'][0]
```

```python
TOTALSTOCK
```

```python
# 2022年底結算總價值/花費的金額
TOTALSTOCK*HDV['2022']['Close'][-1]/TOTALCASH
```

```python
HDV['2022']['Close'].plot()
```
